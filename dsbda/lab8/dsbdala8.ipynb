```python
#null values
print("total number of null values in dataset:",df.isna().sum())
df['Age'].fillna(df['Age'].mean(),inplace=True)
print("")
print("total number of null values in dataset:",df.isna().sum())
```

```python
#one variable histogram
fig, axis = plt.subplots(1, 3)
sns.histplot(ax=axis[0], data=df, x='Sex', hue='Sex', multiple='dodge', shrink=1)
sns.histplot(ax=axis[1], data=df, x='Pclass', hue='Pclass', multiple='dodge', shrink=1)
sns.histplot(ax=axis[2], data=df, x='Survived', hue='Survived', multiple='dodge', shrink=0.8)
plt.show()
```

```python
#one variable histogram
#using kde for better distribution and also as kde is used in case of continous varibles and hue in case of categorial variables
fig,axis=plt.subplots(1,2)
sns.histplot(ax=axis[0],data=df,x='Age',multiple='dodge',shrink=0.8,kde=True)
sns.histplot(ax=axis[1],data=df,x='Fare',multiple='dodge',shrink=0.8,kde=True)
plt.show()
```

```python
#two variable histogram
fig,axis=plt.subplots(2,2)
sns.histplot(ax=axis[0][0],data=df,x='Fare',hue='Sex',multiple='dodge',shrink=1,kde=True)
sns.histplot(ax=axis[0][1],data=df,x='Age',hue='Sex',multiple='dodge',shrink=1,kde=True)
sns.histplot(ax=axis[1][0],data=df,x='Fare',hue='Survived',multiple='dodge',shrink=1,kde=True)
sns.histplot(ax=axis[1][1],data=df,x='Age',hue='Survived',multiple='dodge',shrink=1,kde=True)
plt.show()
```

```python
fig, axis = plt.subplots(2,2)

sns.histplot(ax=axis[0,0], data=df, x='Sex', hue='Survived', multiple= 'dodge', shrink = 0.8, kde = True)
sns.histplot(ax=axis[0,1], data=df, x='Pclass', hue='Survived', multiple='dodge', shrink=0.8, kde= True)
sns.histplot(ax=axis[1,0], data=df, x='Age', hue='Survived', multiple='dodge', shrink = 0.8, kde = True)
sns.histplot(ax=axis[1,1], data=df, x='Fare', hue='Survived', multiple='dodge', shrink = 0.8, kde = True)

plt.show()

```
