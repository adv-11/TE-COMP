```python
#One variable
fig, axes = plt.subplots(1,2)
sns.boxplot(data = df, x ='Age', ax=axes[0])
sns.boxplot(data = df, x ='Fare', ax=axes[1])
plt.show()
```

```python
#One variable
fig, axes = plt.subplots(1,2)
sns.boxplot(data = df, y ='Age', ax=axes[0])
sns.boxplot(data = df, y ='Fare', ax=axes[1])
plt.show()
```

```python
#two variables
fig,axes=plt.subplots(1,3,  sharey=True)
sns.boxplot(ax=axes[0],x='Sex',y='Age',hue='Sex',data=df)
sns.boxplot(ax=axes[1],x='Pclass',y='Age',hue='Pclass',data=df)
sns.boxplot(ax=axes[2],x='Survived',y='Age',hue='Survived',data=df)
plt.show()
```

```python
 #three variables
 fig, axes = plt.subplots(1,2, sharey=True)
 sns.boxplot(ax=axes[0],data = df, x='Sex', y ='Age', hue = 'Survived')
 sns.boxplot( ax=axes[1],data = df, x='Pclass', y ='Age', hue = 'Survived')
 plt.show()
```
